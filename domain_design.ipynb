```python
import matplotlib.pyplot as plt
import proplot as pplt
import numpy as np
import netCDF4 as nc
```

```python
fh = nc.Dataset('TROPESS_reanalysis_mon_emi_nox_anth_2021.nc')
print(fh.variables.keys())
lons = fh.variables['lon'][:]
lats = fh.variables['lat'][:]
fh.close()
```

```python
longrid, latgrid = np.meshgrid(lons, lats)
```

```python
fig = pplt.figure(refwidth=10)
axs = fig.subplots(nrows=1, proj='robin', proj_kw={'lon_0': 180})
# proj = pplt.Proj('robin', lon_0=180)
# axs = pplt.subplots(nrows=2, proj=proj)  # equivalent to above
axs.scatter(longrid[90:146, 165:285].flatten(), latgrid[90:146, 165:285].flatten(), markersize=1)
axs.format(
    suptitle='Figure with single projection',
    latlines=30, lonlines=60, coast=True
)
```
